```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import euclidean_distances
```

```python
def update_coor_mov(coor, mov):
    coor = coor + mov
    upper = coor > 1
    lower = coor < 0
    coor[upper] = 2 - coor[upper] 
    coor[lower] = - coor[lower]
    mov[upper] = mov[upper] * -1
    mov[lower] = mov[lower] * -1
    
    return coor, mov

# def update_day():
    
```

```python
n = 100
status = np.zeros(n)
status[0] = 1
angle = np.random.uniform(0, 2 * np.pi, n)
speed = np.random.uniform(0.5, 1, n)
comp = np.exp(angle * 1j) * speed
mov = np.vstack((comp.real, comp.imag)).transpose(1, 0) / 100
coor = np.random.uniform(size=(n, 2))

fig_size = 6
fig, ax = plt.subplots(figsize=(fig_size, fig_size))
plt.xlim(0, 1)
plt.ylim(0, 1)
for i in range(10):
    g_coor = coor[status == 0]
    y_coor = coor[status == 1]
    r_coor = coor[status > 0]
    ax.scatter(g_coor[:, 0], g_coor[:, 1], c='g')
    ax.scatter(y_coor[:, 0], y_coor[:, 1], s=300, c='y', alpha=0.75)
    ax.scatter(r_coor[:, 0], r_coor[:, 1], c='r')
    coor, mov = update_coor_mov(coor, mov)
    status[status !=0 ] += 1
    status[i] = 1
```
